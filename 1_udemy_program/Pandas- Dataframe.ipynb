# .apply() method on single column - 

```python
def encodeSalary(salary):
    if salary == ' <=50K':
        return 1
    else:
        return 2

df['salary'].apply(encodeSalary)
```

```python
df.head(10)
```

```python
def encodeSalary(salary):
    if salary == ' <=50K':
        return 1
    else:
        return 2

df['encoded_salary'] = df['salary'].apply(encodeSalary)

df.head(10)
```

# .apply() method on multiple columns - 

```python
df.head(10)
```

```python
def desc(list):
    sex = list[9]
    salary = list[14]
    temp = sex + ' earning ' + salary + '$'
    return temp

df.apply(func = desc, axis = 'columns')
```

```python
df['Description'] = df.apply(func = desc, axis = 'columns')

df.head(10)
```
